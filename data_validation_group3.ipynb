# 1. Instalasi dan Persiapan Context

```python
# Instalasi Great Expectations
!pip install -q "great-expectations==0.18.19"

import pandas as pd
from great_expectations.data_context import FileDataContext
from great_expectations.dataset import PandasDataset

# 1. Inisialisasi Context (Membuat folder gx/)
context = FileDataContext.create(project_root_dir='./')
```

# 2. Menghubungkan Data (Datasource & Data Asset)

```python
# 2. Load data asli dan ubah menjadi PandasDataset untuk validasi
df_raw = pd.read_csv('transform_result_digital_marketing_campaign_pipeline.csv')
df_gx = PandasDataset(df_raw)

# 3. Konfigurasi Datasource
datasource_name = 'marketing-datasource'
if datasource_name in context.datasources:
    context.delete_datasource(datasource_name)

datasource = context.sources.add_pandas(name=datasource_name)

# 4. Tambahkan Data Asset dan Batch Request
asset_name = 'digital-marketing-campaign'
asset = datasource.add_dataframe_asset(name=asset_name) 
batch_request = asset.build_batch_request(dataframe=df_gx)
```

# 3. Membuat Expectation Suite dan Validator

```python
# 5. Buat Expectation Suite
expectation_suite_name = 'marketing-validation-suite'
context.add_or_update_expectation_suite(expectation_suite_name)

# 6. Membuat Validator
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)
```

## 1. Validasi Kolom CustomerID dan Conversion (Variabel Utama)

```python
# validasi CustomerID tidak null
validator.expect_column_values_to_not_be_null('CustomerID')
```

```python
# Memastikan CustomerID unik
validator.expect_column_to_exist("Conversion")
```

## 2. Validasi Kolom Age (Usia)

```python
# Memastikan Age berada di rentang 18 sampai 70 tahun
validator.expect_column_values_to_be_between('Age', min_value=18, max_value=70)
```

## 3. Validasi Kolom Gender

```python
# Memastikan nilai Gender hanya 'Male' atau 'Female'
validator.expect_column_values_to_be_in_set('Gender', ['Male', 'Female'])
```

## 4. Validasi Kolom CampaignChannel

```python
# Memastikan CampaignChannel sesuai dengan kategori yang ditentukan
validator.expect_column_values_to_be_in_set(
    'CampaignChannel', 
    ['Social Media', 'Email', 'PPC', 'Referral', 'SEO']
)
```

## 5. Validasi Kolom AdSpend (Biaya Iklan)

```python
# Memastikan biaya iklan minimal 0
validator.expect_column_values_to_be_between('AdSpend', min_value=0)
```

## 6. Validasi Kolom ClickThroughRate (CTR)

```python
# Memastikan CTR berada di rentang 0 sampai 1
validator.expect_column_values_to_be_between('ClickThroughRate', min_value=0, max_value=1)
```

## 7. Validasi Tipe Data

```python
# Memastikan CustomerID bertipe integer
validator.expect_column_values_to_be_of_type('AdSpend', 'float')
```

# Menyimpan Expectation Suite

```python
# Simpan hasil konfigurasi validator ke dalam suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

# Menjalankan Checkpoint

```python
# 1. Tentukan nama checkpoint
checkpoint_name = 'marketing_checkpoint'

# 2. Konfigurasi Checkpoint
checkpoint_config = {
    "name": checkpoint_name,
    "validations": [
        {
            "batch_request": batch_request,
            "expectation_suite_name": expectation_suite_name
        }
    ]
}

# 3. Add or Update Checkpoint ke dalam context
context.add_or_update_checkpoint(**checkpoint_config)

# 4. Jalankan Checkpoint
checkpoint_result = context.run_checkpoint(checkpoint_name=checkpoint_name)

# 5. Menampilkan hasil ringkas
print(f"Checkpoint '{checkpoint_name}': {'SUCCESS' if checkpoint_result.success else 'FAILED'}")
```
